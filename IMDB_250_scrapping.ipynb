```python
import requests
from bs4 import BeautifulSoup
```

```python
response=requests.get("https://www.imdb.com/chart/top/",headers={"User-Agent":"Mozilla/5.0"})
print(response)
```

```python
soup=BeautifulSoup(response.text,"html.parser")
```

```python
data=soup.find("script",id="__NEXT_DATA__")
print(data)
```

```python
print(data)
```

```python
data=eval(data)
print(type(data))
```

```python
data_fetched = data['props']['pageProps']['pageData']['chartTitles']['edges']
print(len(data_fetched))
```

```python
print(data_fetched)
```

```python
movie_name = data_fetched[0]['node']['titleText']['text']
print(movie_name)
```

```python
year=data_fetched[0]['node']['releaseYear']['year']
print(year)
```

```python
ratings=data_fetched[0]['node']['ratingsSummary']['aggregateRating']
print(ratings)
```

```python
votes=data_fetched[0]['node']['ratingsSummary']['voteCount']
print(votes)
```

```python
import pandas as pd

l1 = []
l2 = []
l3 = []
l4 = []

for i in range(len(data_fetched)):
    movie_name = data_fetched[i]['node']['titleText']['text']
    year = data_fetched[i]['node']['releaseYear']['year']
    ratings = data_fetched[i]['node']['ratingsSummary']['aggregateRating']
    votes = data_fetched[i]['node']['ratingsSummary']['voteCount']
    l1.append(movie_name)
    l2.append(year)
    l3.append(ratings)
    l4.append(votes)

# Create DataFrame
d = {'movie_name': l1, 'year': l2, 'ratings': l3, 'votes': l4}
df = pd.DataFrame(d)
print(df)
```

```python
df.to_csv("top250Movies.csv")
```
